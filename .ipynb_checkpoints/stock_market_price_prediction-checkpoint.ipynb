# Stock Market Price Prediction

## Importing the Library

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Importing the Dataset

```python
dataset = pd.read_csv("S&P500_Stock_Data.csv")
```

```python
dataset
```

```python
dataset.describe()
```

## Visualizing the Dataset

```python
sns.pairplot(data=dataset)
```

## Creating the Testing and Training Dataset

```python
X = dataset.iloc[:, :-1]
y = dataset.iloc[:, -1:]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

## Training the Model

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
print("Linear Model Coefficient (m)", regressor.coef_)
print("Linear Model Coefficient (b)", regressor.intercept_)
```

## Evaluating the Model

```python
y_pred = regressor.predict(X_test)
```

```python
k = X_test.shape[1]
n = len(X_test)
```

```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
from math import sqrt
RMSE = float(format(np.sqrt(mean_squared_error(y_test, y_pred)), '.3f'))
MSE = mean_squared_error(y_test, y_pred)
MAE = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
adj_r2 = 1 - (1 - r2) * (n - 1) / (n - k - 1)
MAPE = np.mean(np.abs((y_test - y_pred) / y_test)) * 100
```

```python
print("RMSE =", RMSE, "\nMSE =", MSE, "\nMAE =", MAE, "\nR2 =", r2, "\nAdjusted R2 =", adj_r2, "\nMAPE =", MAPE)
```

## Visualizing the Results

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
x_surf, y_surf = np.meshgrid(np.linspace(dataset["Interest Rates"].min(), dataset["Interest Rates"].max(), 100),
                             np.linspace(dataset["Employment"].min(), dataset["Employment"].max(), 100))
```

```python
only_x = pd.DataFrame({"Interest Rates": x_surf.ravel(), "Employment": y_surf.ravel()})
```

```python
y_pred = regressor.predict(only_x)
```

```python
y_pred = y_pred.reshape(x_surf.shape)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection="3d")
ax.scatter(dataset["Interest Rates"], dataset["Employment"], dataset["S&P 500 Price"], c="blue", marker="x")
ax.plot_surface(x_surf, y_surf, y_pred, color="red", alpha=0.3)
ax.set_xlabel("Interest Rates")
ax.set_ylabel("Employment")
ax.set_zlabel("S&P 500 Price")
ax.view_init(10, 40)
```
